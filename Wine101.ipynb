### Unique flavors

```python
uniqueFlavors = []

for flavor in set(dfWine["Flavors"]):
    for f in flavor.split(","):
        f = f.strip()
        if f not in uniqueFlavors:
            uniqueFlavors.append(f)
```

```python
print("We have " + str(len(uniqueFlavors)) +" unique flavors")
```
